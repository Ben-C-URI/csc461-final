```python
# basic imports

import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import torch


# load data

cowCodes = pd.read_csv('data/COW-country-codes.csv')
sct_data = pd.read_csv('data/sct_df_edited.csv')
cirights = pd.read_excel('data/cirights_v3.12.10.24.xlsx')
```

```python
cirights['ciri'].describe()
```

```python
cirights.rename(columns={'cow':'ccode'}, inplace=True)
cirights = cirights[['ccode', 'year', 'ciri']]
```

```python
df = pd.merge(sct_data, cirights, on=['ccode', 'year'], how='left', suffixes=('_sct', '_cirights'))

df = df.dropna(subset=['ciri'])

df
```

```python
# Filter sct_data to 'inflation', 'gdpgrowth', 'log_itpop', 'region', 'imilex', 'coup', 'coup_lag', 'bankingcrisis', 'systemiccrisis', 'goldstandard', 'domestic_debt_in_default', 'independence', 'gwf_autocracy', 'gwf_military', 'gwf_monarchy', 'gwf_party', 'gwf_personal', 'gwf_democracy'

df = df[['year', 'gwf_country', 'ccode', 'region', 'ben_impos', 'ben_threat',  'ongoing_sanction',  'inflation', 'gdpgrowth', 
                     'log_itpop', 'region', 'imilex', 'coup', 'coup_lag', 'bankingcrisis', 
                     'systemiccrisis', 'goldstandard', 'domestic_debt_in_default', 'independence', 
                     'gwf_autocracy', 'gwf_military', 'gwf_monarchy', 'gwf_party', 'gwf_personal', 'gwf_democracy', 'ciri']]
```
